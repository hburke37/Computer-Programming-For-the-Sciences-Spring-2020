## example with storage

```python
# parameters

k=2
m=1

# initial values

x=0
v=3
t=0

# time parameters
dt=0.001
max_time=10
number_of_steps=int(max_time/dt)

t_list=[]
x_list=[]
v_list=[]

for i in range(number_of_steps):
    
    # changes
    dx=(v)*dt
    dv=(-k*x/m)*dt
    
    #update values
    t=t+dt
    x=x+dx
    v=v+dv
    
    #saving the values
    t_list.append(t)
    x_list.append(x)
    v_list.append(v)
    
    
t=array(t_list)
x=array(x_list)
v=array(v_list)
    




```

```python
subplot(2,1,1)
plot(t,x)
ylabel('position')

subplot(2,1,2)
plot(t,v)

plot(t,2*v)
ylabel('velocity')
xlabel('time')
```

```python
# parameters

k=20
m=1

# initial values

x=0
v=3
t=0

# time parameters
dt=0.001
max_time=100
number_of_steps=int(max_time/dt)

S=Storage()

for i in range(number_of_steps):
    
    # changes
    dx=(v)*dt
    dv=(-k*x/m)*dt
    
    #update values
    t=t+dt
    x=x+dx
    v=v+dv
    
    #saving the values
    S+=t,x,v
    
    
t,x,v=S.arrays()    




```

```python
subplot(2,1,1)
plot(t,x)
ylabel('position')

subplot(2,1,2)
plot(t,v,'r-')
ylabel('velocity')
xlabel('time')
```

```python
# parameters

k=20
m=1

# initial values

variables=array([0,3])
# x=0
# v=3
t=0

# time parameters
dt=0.001
max_time=100
number_of_steps=int(max_time/dt)

S=Storage()

for i in range(number_of_steps):
    
    # changes
    x,v=variables
    change_in_variables=array([v,-k*x/m])
#     dx=(v)*dt
#     dv=(-k*x/m)*dt
    
    #update values
    t=t+dt
    variables=variables+change_in_variables*dt
#     x=x+dx
#     v=v+dv
    
    #saving the values
    x,v=variables
    S+=t,x,v
    
    
t,x,v=S.arrays()    




```

```python
subplot(2,1,1)
plot(t,x)
ylabel('position')

subplot(2,1,2)
plot(t,v,'r-')
ylabel('velocity')
xlabel('time')
```

# infinite integers

```python
def factorial(N):
    
    value=1
    for i in range(1,N+1):
        value=value*i
        
    return value
```

```python
factorial(5)
```

```python
factorial(70)
```

```python
factorial(700)
```
